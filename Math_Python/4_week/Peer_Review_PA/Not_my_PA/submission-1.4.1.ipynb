# Центральная предельная теорема своими руками

```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
%matplotlib inline
```

https://ru.wikipedia.org/wiki/Гамма-распределение

```python
k=9
th=0.5
gamma = sts.gamma(k,0,th) #Гамма распредление с параметрами k,th.
sample = gamma.rvs(1000) #Генерируем выборку.
```

```python
x = np.linspace(-5,20,1000)
pdf = gamma.pdf(x)

plt.plot(x,pdf,'r',label='Gamma pdf') #Строим график реальной плотности гамма распределения.
plt.hist(sample,normed=True) #Строим гистрограмму из выборки.
plt.legend(loc='upper right')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
#Вычисление теоретических мат.ожидания и дисперсии.
EX = k*th
DX = k*th**2
```

```python
#Генерация 1000 выборок объемом 5,10,50 и построение гистрограмм распредления их выборочных средних.
n = [5,10,50]
for N in n:
    samplen = []
    norm = sts.norm(EX,sqrt(DX/float(N)))#Пересчет EX и DX для норм.распр из EX и DX исходного распредления.
    pdf = norm.pdf(x)
    plt.plot(x,pdf) #Наложение графиков плотностей нормальных распределений на исходные гистограммы.
    for i in xrange(1001):
        samplen.append(sum(gamma.rvs(N))/N)
    plt.hist(samplen,normed=True, label='$N=%s$' % N)#Построение гистрограмм.
    plt.legend(loc='upper right')
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.show()
```

На примерах построенных графиков видно, насколько хорошо выполняется ЦПТ. Так же с помощью этих графиков легко сделать вывод, что точность аппроксимации возрастает с ростом N.
