```python
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
```

```python
lines = []
data = []

with open('checkins.csv') as input_file:
    columns = map(lambda x: x.strip(), input_file.readline().strip().split('|'))
    n_params = len(columns)
    for line in input_file:
        items = map(lambda x: x.strip(), line.strip().split('|'))
        if len(items) != n_params:
            print line
        else:
            data.append(items)
```

```python
data
```

```python
df = pd.DataFrame(data)
df.columns = columns
df = df[(df.latitude != '') & (df.longitude != '')]
df.head()
```

```python
df_coord = df[['latitude','longitude']]
df_coord = df_coord.applymap(float)
```

```python
data_sample = df_coord.sample(100000)
```

```python
MS = MeanShift(bandwidth = 0.1)
```

```python
MS.fit(data_sample)
```

```python
labels = MS.labels_
cluster_centers = MS.cluster_centers_
labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("number of estimated clusters : %d" % n_clusters_)
```

```python
preds = MS.predict(data_sample)
```

```python
chosen_clusters_centers = []
for i in range(len(N_cluster)):
    chosen_clusters_centers.append(cluster_centers[N_cluster[i]])
chosen_clusters_centers = np.array(chosen_clusters_centers)
```

```python
def dist(x,y):   
    return np.sqrt(np.sum((x-y)**2))
dist(chosen_clusters_centers[609],offices[0])
```

```python
distances = []
for i in range(chosen_clusters_centers.shape[0]):
    helping_list = []
    for j in range(offices.shape[0]):
        helping_list.append(dist(chosen_clusters_centers[i],offices[j]))
        if j == 5:
            distances.append(sorted(helping_list)[0])
```

```python
z = 0
min = 1
for i in range(len(distances)):
    if min > distances[i]:
        min = distances[i]
        z = i
```

```python
best_coordinates = chosen_clusters_centers[z]
```
